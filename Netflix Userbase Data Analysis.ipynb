# Country And Count

```python
plt.figure(figsize=(12,6))
A = sns.countplot(x="Country",data=df)
for bars in A.containers:
    A.bar_label(bars)
```

```python
# Country name with Subscription Count and united states and spain has max users are thier.
```

# Gender Vs Country

```python
plt.figure(figsize=(14,6))
sns.countplot(x='Country',hue='Gender',data=df)
plt.title("Gender Distribution By Country")
plt.show()
```

```python
#Most of  female,male user in us and Spain have most female user as compaire to male. 
```

# Subscription Type Vs Gender

```python
plt.figure(figsize=(6,4))
sns.countplot(df,x="Subscription Type",hue="Gender")
plt.title("Count of Subscription Type base on Gender")
plt.show()
```

```python
#Female have 500 basic subscription type as compaire male have only  490. 
#and in premium Subscription type male have 360 and female have only 350.In Standard male 380 and female have 370 count.
```

# Monthly Revenue Basis on Country

```python
Counrty_revenue= df.groupby(["Country"],as_index=False)["Monthly Revenue"].sum().sort_values(by="Monthly Revenue",ascending=False).head()
sns.set(rc={"figure.figsize":(12,4)})
sns.barplot(data =Counrty_revenue ,x="Country",y="Monthly Revenue")
plt.show()
```

```python
# Most Montly revenue is in us,spain and canada as compair to other on the basis on bar.
```
